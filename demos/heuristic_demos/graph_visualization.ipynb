```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
from pathlib import Path

data = "Graph & Nodes & Edges & Clustering & Diameter & Density \\\\ \n"

for file in Path('graphs').glob('*.txt'):
    G = nx.read_graphml(file)
    print(file)
    print(f"Number of nodes: {len(G.nodes())}")
    print(f"Number of edges: {len(G.edges())}")
    average_clustering = nx.average_clustering(G)
    print(f"Average Clustering Coefficient: {average_clustering}")
    diameter = nx.diameter(G)
    print(f"Graph Diameter: {diameter}")
    density = nx.density(G)
    print(f"Graph Density: {density}")

    data += f"{file.name} & {G.number_of_nodes()} & {G.number_of_edges()} & {average_clustering} & {diameter} & {density} \\\\ \n"


with open(f"graphs/data.csv", 'w') as f:
    f.write(data)
```
